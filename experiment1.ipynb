## Website visitor converion simulation experiments

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
sns.set(style="darkgrid")
```

```python
visitors1 = 100  # number of visitors per timestep
conversion1 = 0.05  # simple conversion rate

timesteps = 100  # number of timesteps to consider
```

### Simple extrapolation

```python
def simple(visitors, conversion, timesteps):
    '''Simple linear calculation of converted visits
    '''
    results = np.zeros((timesteps, 3))
    results[:, 0] = np.arange(0, timesteps)
    for i in range(0, timesteps):
        results[i, 1] = np.round(visitors * conversion)
        if i > 0:
            results[i, 2] = results[i-1, 2] + results[i, 1]
        else:
            results[i, 2] = results[i, 1]
    return pd.DataFrame(results, columns=["time", "conversion", "cummulative_conversion"])

results1 = simple(visitors1, conversion1, timesteps)
print("Calculated results")
print(results1)
sns.relplot(x="time", y="cummulative_conversion", kind="line", data=results1)
```

### Probabilistic conversion with the same settings

```python
# Set random seed
np.random.seed(0)

def probabilistic(visitors, conversion, timesteps):
    '''Probabilistic conversion, random conversion
    rates of visitors with conversion rate.
    '''
    results = np.zeros((timesteps, 3))
    results[:, 0] = np.arange(0, timesteps)
    
    distribution = stats.bernoulli(conversion)
    
    for i in range(0, timesteps):
        results[i, 1] = np.sum(distribution.rvs(np.int(visitors)))
        if i > 0:
            results[i, 2] = results[i-1, 2] + results[i, 1]
        else:
            results[i, 2] = results[i, 1]
    
    return pd.DataFrame(results, columns=["time", "conversion", "cummulative_conversion"])

results2 = probabilistic(visitors1, conversion1, timesteps)
print("Calculated results")
print(results2)
sns.relplot(x="time", y="cummulative_conversion", kind="line", data=results2)
```

```python
# reseting the random variable so the example calculation will start the same as above,
# so that the same run is included in the results below as well
np.random.seed(0)

repeats = 10  # how many repetitions to run
f, axes = plt.subplots(2, 1, figsize=(8, 8))
finals = np.zeros(repeats)  # collect the final converted count

for i in range(repeats):
    res = probabilistic(visitors1, conversion1, timesteps)
    finals[i] = res['cummulative_conversion'].tail(1)
    sns.lineplot(x="time", y="cummulative_conversion", data=res, ax=axes[0])

total_conversion = pd.Series(finals, name="total conversions")
sns.distplot(total_conversion, ax=axes[1]);
print("Mean: {0}, Std: {1:.4}".format(np.mean(finals), np.std(finals)))
```

```python
np.random.seed(0)

repeats = 1000
f, axes = plt.subplots(2, 1, figsize=(7, 7))
finals = np.zeros(repeats)

for i in range(repeats):
    if i % 100 == 0:
        # To show progress
        print("Simulation: {}".format(i))
    res = probabilistic(visitors1, conversion1, timesteps)
    finals[i] = res['cummulative_conversion'].tail(1)
    sns.lineplot(x="time", y="cummulative_conversion", data=res, ax=axes[0])

total_conversion = pd.Series(finals, name="total conversions")
sns.distplot(total_conversion, ax=axes[1]);

print("Mean: {0}, Std: {1:.4}".format(np.mean(finals), np.std(finals)))
```

### Full probabilistic simulation

```python
visitors_std = 20
conversion_std = 0.01

visitors2 = stats.norm(visitors1, visitors_std)
conversion2 = stats.norm(conversion1, conversion_std)
```

```python
np.random.seed(0)

def probabilistic_dist(visitors, conversion, timesteps):
    results = np.zeros((timesteps, 4))
    results[:, 0] = np.arange(0, timesteps)
    for i in range(0, timesteps):
        conversion_now = stats.bernoulli(conversion.rvs())
        visitors_now = np.int(np.round(visitors.rvs()))
        results[i, 1] = visitors_now
        results[i, 2] = np.sum(conversion_now.rvs(visitors_now))
        if i > 0:
            results[i, 3] = results[i-1, 3] + results[i, 2]
        else:
            results[i, 3] = results[i, 2]
    return pd.DataFrame(results, columns=["time", "visitors", "conversion", "cummulative_conversion"])


results_mc1 = probabilistic_dist(visitors2, conversion2, timesteps)
print(results_mc1)
sns.relplot(x="time", y="cummulative_conversion", kind="line", data=results_mc1)
```

```python
np.random.seed(0)

repeats = 1000
f, axes = plt.subplots(2, 1, figsize=(7, 7))
v = visitors2.rvs(repeats)
c = conversion2.rvs(repeats)
finals = np.zeros(repeats)

for i in range(repeats):
    if i % 100 == 0:
        # To show progress
        print("Simulation: {}".format(i))
    res = probabilistic_dist(visitors2, conversion2, timesteps)
    finals[i] = res['cummulative_conversion'].tail(1)
    sns.lineplot(x="time", y="cummulative_conversion", data=res, ax=axes[0])

total_conversion = pd.Series(finals, name="total conversions")
sns.distplot(total_conversion, ax=axes[1]);

print("Mean: {0}, Std: {1:.4}".format(np.mean(finals), np.std(finals)))
```

### Churn model

```python
# Set a rate of churn, while keeping everything else the same
churn = stats.norm(0.005, 0.001)
initial_users = 0

np.random.seed(0)

def probabilistic_dist2(visitors, conversion, churn, initial_users, timesteps):
    results = np.zeros((timesteps, 5))
    results[:, 0] = np.arange(0, timesteps)
    for i in range(0, timesteps):
        conversion_now = stats.bernoulli(conversion.rvs())
        visitors_now = np.int(np.round(visitors.rvs()))
        churn_now = stats.bernoulli(churn.rvs())
        results[i, 1] = visitors_now
        results[i, 2] = np.sum(conversion_now.rvs(visitors_now))
        if i > 0:
            churn_count = np.sum(churn_now.rvs(np.int(results[i-1, 4])))
            results[i, 3] = churn_count
            results[i, 4] = results[i-1, 4] + results[i, 2] - results[i, 3]
        else:
            churn_count = np.sum(churn_now.rvs(initial_users))
            results[i, 3] = churn_count
            results[i, 4] = results[i, 2] + initial_users
    return pd.DataFrame(results, columns=["time", "visitors", "conversion", "churn", "cummulative_conversion"])


results_mc1 = probabilistic_dist2(visitors2, conversion2, churn, initial_users, timesteps)
print(results_mc1)
sns.relplot(x="time", y="cummulative_conversion", kind="line", data=results_mc1)
```

```python
np.random.seed(0)

repeats = 1000
f, axes = plt.subplots(2, 1, figsize=(7, 7))
v = visitors2.rvs(repeats)
c = conversion2.rvs(repeats)
finals = np.zeros(repeats)

for i in range(repeats):
    if i % 100 == 0:
        # To show progress
        print("Simulation: {}".format(i))
    res = probabilistic_dist2(visitors2, conversion2, churn, initial_users, timesteps)
    finals[i] = res['cummulative_conversion'].tail(1)
    sns.lineplot(x="time", y="cummulative_conversion", data=res, ax=axes[0])

total_conversion = pd.Series(finals, name="total conversions")
sns.distplot(total_conversion, ax=axes[1]);

print("Mean: {0}, Std: {1:.4}".format(np.mean(finals), np.std(finals)))
```
